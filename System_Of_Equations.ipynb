```python
import numpy as np

# Problem 1
# STEP 1: Define the coefficient matrix (A)
# Each row represents one of the three intercepted signal relationships.
# Row 1: 2(m1) + 1(m2) + 1(m3) = 36
# Row 2: -1(m1) + 2(m2) - 1(m3) = -11
# Row 3: 3(m1) - 1(m2) + 2(m3) = 47
A = np.array([
    [2, 1, 1],
    [-1, 2, -1],
    [3, -1, 2]
])

# STEP 2: Define the constant vector (b)
# This represents the intercepted output values (the results of the equations).
b = np.array([36, -11, 47])

# STEP 3: Attempt to decrypt the message.
# We try to find the original packets (m) by calculating m = A^-1 * b.
# NOTE: Because Eq 3 is just Eq 1 minus Eq 2, this is a "singular matrix" 
# (determinant is 0). We use a try-except block to handle the mathematical error gracefully.
try:
    m = np.linalg.inv(A) @ b
    print("Decrypted Message Packets (m1, m2, m3):")
    print(m)
except np.linalg.LinAlgError as e:
    print("Execution Error:", e)
    print("Mathematical Conclusion: The matrix is singular because the equations are linearly dependent.")
    print("There is no unique inverse, meaning there are infinite possible decryption keys.")
```

```python
import numpy as np

# Problem 2
# STEP 1: Define the nutrient matrix (A)
# Each column represents a Pellet (A, B, C, D)
# Each row represents a nutrient (Protein, Carbs, Fat, Vitamins)
A = np.array([
    [30, 20, 50, 10], # Protein content
    [40, 50, 20, 30], # Carbohydrate content
    [10, 30, 20, 40], # Fat content
    [20, 10, 40, 50]  # Vitamin content
])

# STEP 2: Define the target nutrient vector (b)
# These are the exact grams/mg the primate needs daily.
b = np.array([145, 210, 115, 170])

# STEP 3: Solve for the number of servings (x)
# The np.linalg.inv(A) computes the inverse matrix.
# The '@' symbol calculates the dot product of the inverse matrix and vector b.
x = np.linalg.inv(A) @ b

print("Required servings of Pellets A, B, C, and D (per 100g):")
print(np.round(x, 2))
```

```python
import numpy as np

# Problem 3
# STEP 1: Define the 3x3 Consumption Matrix (C)
# Columns represent the sector providing the input (E, M, S)
# Rows represent the sector receiving the input
C = np.array([
    [0.2, 0.3, 0.1], # Inputs required from Energy
    [0.4, 0.1, 0.2], # Inputs required from Manufacturing
    [0.1, 0.2, 0.1]  # Inputs required from Services
])

# STEP 2: Create the 3x3 Identity Matrix (I)
# np.eye(3) automatically generates a 3x3 matrix with 1s on the diagonal and 0s elsewhere.
I = np.eye(3)

# STEP 3: Define the external demand vector (d) in millions
d = np.array([500, 800, 600])

# STEP 4: Compute the matrix (I - C)
# This subtracts the consumption matrix from the identity matrix element-wise.
A_leontief = I - C

# STEP 5: Solve for total production (x)
# We find the inverse of our new (I-C) matrix and multiply it by demand (d).
x_production = np.linalg.inv(A_leontief) @ d

print("Total Production Required (in millions) for Energy, Manufacturing, and Services:")
print(np.round(x_production, 2))
```

```python
import numpy as np

# Problem 4
# STEP 1: Assemble the 4x4 Resistance Matrix (R)
# Each row represents a loop equation. 
# We explicitly place a 0 wherever a current (I) is not part of that loop.
R = np.array([
    [10, -2, -3,  0],  # Loop 1
    [-2, 15,  0, -4],  # Loop 2
    [-3,  0, 12, -5],  # Loop 3
    [ 0, -4, -5, 20]   # Loop 4
])

# STEP 2: Define the Voltage Vector (V)
# This holds the constant voltage sums from the right side of the equations.
V = np.array([12, 5, 0, -10])

# STEP 3: Solve for the Currents (I)
# The inverse of the resistance matrix multiplied by the voltages gives the exact current flows.
currents = np.linalg.inv(R) @ V

print("Currents flowing through Loops I1, I2, I3, and I4 (in Amperes):")
print(np.round(currents, 4))
```

```python
import numpy as np

# Problem 5
# STEP 1: Define the recorded time intervals (x) and altitudes (y) as arrays
x_vals = np.array([-2, -1, 1, 2, 3])
y_vals = np.array([59, 17, 5, 11, 49])

# STEP 2: Construct the 5x5 Vandermonde Matrix (A)
# We use a Python list comprehension to iterate through our x_vals.
# For each x, it calculates the row: [x^0, x^1, x^2, x^3, x^4] automatically.
A = np.array([
    [x**0, x**1, x**2, x**3, x**4] for x in x_vals
])

# STEP 3: Solve for the polynomial coefficients (a, b, c, d, e)
# We invert the Vandermonde matrix and multiply it by the column vector of y-values.
coefficients = np.linalg.inv(A) @ y_vals

print("Polynomial Coefficients (a, b, c, d, e):")
print(np.round(coefficients, 2))
```
